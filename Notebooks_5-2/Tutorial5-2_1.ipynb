```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
import pylab as pl
```

```python
# The 2-dimensional nonlinear system.
def dx_dt(x, t):
    return [a*x[0]+b*x[1],c*x[0]+d*x[1]]
```

```python
# Trajectories in forward time.
a = 1.
b = 1.
c = -1.
d = 1.
ts = np.linspace(0, 10, 500)
ic = np.linspace(-2, 2, 6)
for r in ic:
    for s in ic:
        x0 = [r, s]
        xs = odeint(dx_dt, x0, ts)
        plt.plot(xs[:, 0], xs[:, 1], 'r-')

# Label the axes and set fontsizes.
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.tick_params(labelsize=15)
plt.xlim(-2, 2)
plt.ylim(-2, 2)

# Plot the vectorfield.
X, Y = np.mgrid[-2:2:20j, -2:2:20j]
u = a*X + b*Y
v = c*X + d*Y 
pl.quiver(X, Y, u, v, color='b')
```

```python
# The 2-dimensional nonlinear system.
def dx_dt(x, t):
    lab = 1.0
    om = 1.0
    x3 = x[0]*(x[0]*x[0]+x[1]*x[1])
    y3 = x[1]*(x[0]*x[0]+x[1]*x[1])
    return [lab*x[0]+om*x[1]-x3,lab*x[1]-om*x[0]-y3]
```

```python
# Trajectories in forward time.
ts = np.linspace(0, 10, 500)
ic = np.linspace(-2, 2, 6)
for r in ic:
    for s in ic:
        x0 = [r, s]
        xs = odeint(dx_dt, x0, ts)
        plt.plot(xs[:, 0], xs[:, 1], 'r-')

# Label the axes and set fontsizes.
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.tick_params(labelsize=15)
plt.xlim(-2, 2)
plt.ylim(-2, 2)

# Plot the vectorfield.
X, Y = np.mgrid[-2:2:20j, -2:2:20j]
lab = 1.0
om = 1.0
X3 = X*(X*X + Y*Y)
Y3 = Y*(X*X + Y*Y)
u = lab*X + om*Y - X3
v = lab*Y - om*X - Y3
pl.quiver(X, Y, u, v, color='b')
```
