```python
#EM  Euler-Maruyama method on Stommel Box Model

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(100)

F = 1.1; mu = 6.2; sigma = 0.2; Xzero = 0.1
T=100; N=2**10; dt = T/N
t=np.linspace(0,T,N+1)

dW=np.sqrt(dt)*np.random.randn(1,N)
W=np.cumsum(dW)

R=1; Dt=R*dt; L=N
Xem=np.zeros(L+1); Xem[0] = Xzero

for j in range(1,L+1):
    Winc=np.sum(dW[0][range(R*(j-1),R*j)])   
    Xem[j]=Xem[j-1]+Dt*(F-Xem[j-1]*(1+mu*(1-Xem[j-1])**2))+sigma*Winc

plt.plot(t,Xem)
plt.xlabel(r'$t$',fontsize=16); plt.ylabel(r'$Y$',fontsize=16,rotation=0)
plt.show()
```
